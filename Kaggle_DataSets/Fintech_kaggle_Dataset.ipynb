```python
'''Accessing the number of rows from the dataframe using LOC function -- LOCATION function

loc() is label based data selecting method which means that we have to pass the name of the row or column which we want to select'''

#1. Selecting data fror users whose age > 70 and zodiac sign = Gemini, using loc :

display(df.loc[(df.age > 70) & (df.zodiac_sign == 'Gemini')])

#2. # updating values of churn to 0, for users whose age > 70 and zodiac sign = Gemini :

'''df.loc[(df.age > 70) & (df.zodiac_sign == 'Gemini') , ['churn']] = 1
display(df.loc[(df.age > 70) & (df.zodiac_sign == 'Gemini')])'''
```

```python
'''Accessing the number of rows from the dataframe using iLOC function

#iloc() is a indexed based selecting method which means that we have to pass integer index in the method to select specific row/column'''

display (df.iloc[4:7,0:4])  #This displays the rows of index 4 to 7 and columns of index 0 to 4

display(df.iloc[[0, 2, 4, 7]]) # selecting 0th, 2th, 4th, and 7th index rows
```

```python
#To check the total number of NAN values in the dataframe, across columns :

df.isna().sum()
```

```python
#Replace NAN values of 'Age' column with mean:

df['age'].fillna(df['age'].mean(), inplace=True)
```

```python
#Replace NAN values of 'rewards_earned' column with median:

df['rewards_earned'].fillna(df['rewards_earned'].median(), inplace=True)
```

```python
#Replace NAN values of 'credit_score' column with median:

df['credit_score'].fillna(df['credit_score'].median(), inplace=True)
```

```python
#Sort the dataframe w.r.t age column, ascending order :

df.sort_values("age")
```

```python
#Group the data using 'Groupby' function & applying aggregate function on it :

import numpy as np

df2 = df.groupby('credit_score')

df2['age'].agg([np.sum, np.mean, np.std])
```

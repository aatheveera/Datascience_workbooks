```python
#Remove unwanted columns from the dataframe :

df.drop(['PassengerId', 'Ticket'], axis=1, inplace=True)
```

```python
#Find the total number of NULL values in the Name, Age & Sex columns

print ("The total number of null values in the NAME column is:", df['Name'].isnull().sum())
print ("The total number of null values in the AGE column is:", df['Age'].isnull().sum())
print ("The total number of null values in the SEX column is:", df['Sex'].isnull().sum())
```

```python
#Get the MODE(most frequent) value of the columns Age & Fare

df[['Age','Fare']].mode()
```

```python
#Create a new column 'AgeCategory' based on the age group of the passengers

def conditions(s):
  if (s['Age'] < 18):
    return "Children"
  elif (s['Age'] > 50):
    return "Old"
  elif (s['Age'] == "NaN"):
    return "MissingData"
  else:
    return "Adult"

df['AgeCategory'] = df.apply(conditions, axis=1)
```

```python
sample_loc = df.loc[(df['AgeCategory'] != 'Adult')]

sample_loc
```
